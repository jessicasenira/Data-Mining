```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('restaurtantscore.csv')
data.head()
```

```python
np.sum(data.isnull())
```

```python
data.describe()
```

```python
data['business_latitude']=data['business_latitude'].fillna((data['business_latitude'].mean()))
data['business_longitude']=data['business_longitude'].fillna((data['business_longitude'].mean()))
data['business_phone_number']=data['business_phone_number'].fillna((data['business_phone_number'].mean()))
data['inspection_score']=data['inspection_score'].fillna((data['inspection_score'].mean()))
```

```python
np.sum(data.isnull())
```

```python
from scipy.stats import mode
```

```python
data['business_postal_code']=data['business_postal_code'].fillna(data['business_postal_code']).dropna().mode()[0]
data['business_location']=data['business_location'].fillna(data['business_location']).dropna().mode()[0]
data['violation_id']=data['violation_id'].fillna(data['violation_id']).dropna().mode()[0]
data['violation_description']=data['violation_description'].fillna(data['violation_description']).dropna().mode()[0]
data['risk_category']=data['risk_category'].fillna(data['risk_category']).dropna().mode()[0]
np.sum(data.isnull())
```

```python
from scipy.stats import stats
z = np.abs(stats.zscore(data._get_numeric_data()))
print(z)
data = data[(z<3).all(axis=1)]
print(data.shape)
```
