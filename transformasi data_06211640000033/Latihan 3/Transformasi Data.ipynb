```python
import pandas as pd
import numpy as np 
```

## Import Dataset

```python
pcademo = pd.read_csv('pcademo.csv')
x = pcademo.iloc[0:51,1:20] #iloc untuk mengambil data di lokasi tertentu
x.tail()
```

## Normalization

```python
from sklearn.preprocessing import normalize
xnorm=normalize(x,norm='l2') #default norm L2, other are L1 and max
xnorm=pd.DataFrame(xnorm) #dataframe agar bentuknya tabel
xnorm.tail()
```

## MinMax 

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(-2,1)) #feature range : nilai hasil yg ingin dibuat
scaler.fit(x)
x_scaled=scaler.transform(x)
x_scaled=pd.DataFrame(x_scaled)
x_scaled.tail()
```

## 3sigmoid 

```python
xsigmoidal=(1-2.718281828**(-xnorm))/(1+2.718281828**(-xnorm))
xsigmoidal=pd.DataFrame(xsigmoidal)
xsigmoidal.tail()
```

## Decimal Scalling 

```python
xdecimal=x/(10**3)
xdecimal=pd.DataFrame(xdecimal)
xdecimal.tail()
```

## Softmax

```python
def softmax(x): #mengubah data dgn rentang 0-1
    return np.exp(x)/np.sum(np.exp(x),axis=0) #rumusnya bikin sendiri
xsoft=softmax(xnorm)
xsoft=pd.DataFrame(xsoft)
print(xsoft)
```

## Feature Selection (mencari skor)

```python
from sklearn.feature_selection import SelectKBest, chi2, f_regression
Xx=x.iloc[0:51,2:20]
y=x.iloc[0:51,1]
selector = SelectKBest(f_regression,k="all").fit(Xx,y) #pilih f_regression, k=all untuk mengetahui variabel mana yg memiliki score paling tinggi
X_new=selector.transform(Xx)
scores=selector.scores_
scores #nilai score dari semua var prediktor, semakin tinggi score semakin layak var tsb untuk masuk ke model, pilih bbrp variabel tertinggi
```
